```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/bank-full.csv', sep=';')

print(df.head())
print(df.shape)
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='y', data=df, palette='Set2')
plt.title("Term Deposit Subscription Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['job'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Job Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=20, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['marital'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['lightblue','lightgreen','pink']
)
plt.title("Marital Status Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='education', hue='y', data=df)
plt.xticks(rotation=45)
plt.title("Education vs Term Deposit Subscription")
plt.show()
```

```python
plt.figure(figsize=(8,6))

# convert categorical columns temporarily
df_temp = df.copy()
for col in df_temp.select_dtypes(include='object').columns:
    df_temp[col] = df_temp[col].astype('category').cat.codes

sns.heatmap(df_temp.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
